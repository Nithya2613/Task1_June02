**4. Identify and Handle Missing Values**

```python
df = df.isnull().sum()
df
```

```python
df = pd.read_csv('/content/car_prices.csv')
#Fix missing text values with 'unknown'
text_cols = ['make', 'model', 'trim', 'body', 'transmission', 'color', 'interior']
for col in text_cols:
    df[col] = df[col].fillna('unknown')

#Fill numeric columns with median
numeric_cols = ['condition', 'odometer']
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
print(df.isnull().sum())
```

**5. Remove Duplicates**

```python
df = df.drop_duplicates()
```

**6. Standardize Text Columns**

```python
text_columns = ['make', 'model', 'trim', 'body', 'transmission', 'state', 'color', 'interior', 'seller']
for col in text_columns:
    df[col] = df[col].str.lower().str.strip()
```

**7. Convert Date Formats to Standard  (dd-mm-yyyy)**

```python
#Convert to datetime safely
df['saledate'] = pd.to_datetime(df['saledate'], errors='coerce')

#Drop any failed date parses (NaT)
df = df[df['saledate'].notnull()]

#Format the datetime
df['saledate'] = df['saledate'].dt.strftime('%d-%m-%Y')
```

```python
df['saledate']
```

**8. Rename Column Headers**

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
```

```python
df.columns
```

**9. Fix Data Types**

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['condition'] = pd.to_numeric(df['condition'], errors='coerce')
df['odometer'] = pd.to_numeric(df['odometer'], errors='coerce')
df['mmr'] = pd.to_numeric(df['mmr'], errors='coerce')
df['sellingprice'] = pd.to_numeric(df['sellingprice'], errors='coerce')
```

**10. Outlier Detection and Removal**

```python
# Remove unrealistic odometer values
df1 = df[(df['odometer'] >= 0) & (df['odometer'] <= 300000)]

# Ensure condition is within valid range
df2= df[(df['condition'] >= 1) & (df['condition'] <= 5)]
```
